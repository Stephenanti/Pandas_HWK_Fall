## 4) How many male billionaires are there compared to the number of female billionares? What percent is that? Do they have a different average wealth?

> **TIP:** The last part uses `groupby`, but the count/percent part does not.
> **TIP:** When I say "average," you can pick what kind of average you use.

```python
df.value_counts('gender')
```

```python
df.value_counts('gender').sum()
```

```python
gender_sum = df.value_counts('gender').sum()
male_sum = 1473
female_sum = 180
```

```python
# Percentage of male billionaires
male_sum/gender_sum * 100
round(male_sum/gender_sum * 100,2)
```

```python
# percentage of female billionaires
female_sum/gender_sum * 100
round(female_sum/gender_sum * 100,2)
```

```python
# average wealth of females and males
df.groupby(by='gender').networthusbillion.mean().sort_values()
```
